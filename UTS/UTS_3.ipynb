```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv('./drive/MyDrive/NLP_Dede_Ulfah/Reviews_Filter.csv')
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
compound_scores = []

for review in data['Text']:
    compound_scores.append(sia.polarity_scores(review)['compound'])

data['Compound_Score'] = compound_scores
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Compound_Score'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review')
plt.title('Distribusi Compound Score')
plt.show()
```

```python
coffee_reviews = data[data['Text'].str.contains('coffee', case=False)]
plt.figure(figsize=(10, 6))
plt.hist(coffee_reviews['Compound_Score'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review')
plt.title('Distribusi Compound Score untuk Review Mengandung "coffee"')
plt.show()
```

```python
price_reviews = data[data['Text'].str.contains('price', case=False)]
plt.figure(figsize=(10, 6))
plt.hist(price_reviews['Compound_Score'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Compound Score')
plt.ylabel('Jumlah Review')
plt.title('Distribusi Compound Score untuk Review Mengandung "price"')
plt.show()
```
