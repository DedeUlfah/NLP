Dalam pidato tersebut, Indonesia mengutuk keras tindakan kekerasan di Gaza dan serangan Israel terhadap Rumah Sakit Al-Ahli, menilai itu sebagai pelanggaran hukum internasional.

Pemerintah Indonesia juga berusaha evakuasi WNI yang terdampak konflik. Lebih lanjut, Indonesia bersama OKI berupaya menghentikan eskalasi kekerasan dan menekankan pentingnya penyelesaian akar permasalahan, yaitu pendudukan Israel atas Palestina.

Mereka mendorong solidaritas global dan mendukung parameter internasional dalam menyelesaikan konflik tersebut. Pidato ini mencerminkan komitmen Indonesia terhadap perdamaian di kawasan tersebut.

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
```

```python
with open('PidatoPresidenJokowi.txt', 'r', encoding='latin-1') as file:
    text = file.read()
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

words = text.split()
cleaned_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]
```

```python
most_common = Counter(cleaned_words).most_common(10)
```

```python
x, y = zip(*most_common)
plt.figure(figsize=(10, 6))
plt.barh(x, y)
plt.title('Most Common Keywords')
plt.xlabel('Frequency')
plt.ylabel('Keywords')
plt.show()
```
